```python
plt.figure(figsize=(5,2))

x1 = range(50)
y1 = [np.cos(value) for value in x1]

plt.plot(x1,y1)
```

```python
plt.figure(figsize=(5,2))

x1 = range(50)
y1 = [np.cos(value) for value in x1]

y2 = [np.sin(value) for value in x1]

plt.plot(x1,y1)
plt.plot(x1,y2)
```

```python
plt.figure(figsize=(8,3))

plt.subplot(1,2,1) #1행2열 첫 번째 도화지 subplot - plot을 둘로 나눔
plt.plot(x1,y1,color='blue',linestyle='dotted')

plt.subplot(1,2,2)
plt.plot(x1,y2,color='orange',linestyle='dashed')
```

```python
x1 = range(100) #RGB 256 color 색상표
y1 = [value for value in x1]

x2 = range(100)
y2 = [value + 50 for value in x2]

plt.plot(x1,y1, c = '#CC33CC',ls='-.',label='Line 1' ) 
plt.plot(x2,y2,c = '#3333FF' ,ls=':',label='Line 2')

plt.text(50,70,'Line 1')
plt.annotate('Line 2',xytext=(30,110),xy=(50,100),arrowprops=dict(facecolor = 'yellow',shrink=0.1))

#xy= 화살표 방향(x,y) x는 50 y는 100 위치
#xytext= Line2 위치
#shrink는 text와 거리


plt.legend(fancybox=True,shadow=True,loc='upper right')

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Chart')

plt.savefig('LineChart.jpg')
```
