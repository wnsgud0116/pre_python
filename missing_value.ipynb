```python
import pandas as pd
import numpy as np
```

```python
arr = np.random.randn(5,5)
```

```python
df = pd.DataFrame(arr,columns=['one','two','three','four','five'],\
            index = ['a','b','c','d','e'])
df
```

```python
df.iloc[1:3,1] = np.nan
df
```

```python
df.iloc[2,2:4] = np.nan
```

```python
df
```

```python
#isnull() - null이면 True. not null이면 False
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum(axis=1)
```

```python
# #null값의 처리
# 1.null이 속한 열이나 칼럼을 삭제..
# 2.null값을 다른 값으로 채우는 방법..
```

```python
#null이 속한 열이나 행을 삭제..=> dropna()
```

```python
df
```

```python
df.dropna() #null값이 속한 row를 삭제
```

```python
df.dropna(axis=1) #null값이 속한 col을 삭제
```

```python
df.dropna(thresh=3) #데이터가 3개 이상있으면 삭제하지 않도록
```

```python
#2. null값에 다른 값을 채우는 방법 => fillna()
```

```python
df
```

```python
df.fillna(0)
```

```python
df.fillna(method='ffill') #바로 앞에 있는 값으로 채우기
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(100)
```

```python
df.fillna()
```
