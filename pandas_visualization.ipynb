```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

```python
drinks = pd.read_csv(url)
drinks.head()
```

```python
#drinks.rename(columns = {'col':'col1'}) #하나씩 바꿀때 
drinks.columns = ['country','beer','spirit','wine','liters','continent'] #전체 바꿀때
drinks.head()
```

```python
drinks.info()
```

```python
drinks.set_index('country',inplace=True) #set_index 해당 칼럼을 index로 놓을때
```

```python
drinks
```

```python
#1. Line Chart
```

```python
drinks['beer'].plot()
```

```python
drinks.plot()
```

```python
drinks.sort_values('beer')['beer'].plot()
```

```python
#2. Bar chart
```

```python
drinks['beer'].plot(kind='bar')
```

```python
plt.figure(figsize=[10,3])
drinks['beer'].sort_values(ascending=False)[:50].plot(kind='bar') #[:50] country number
```

```python
drinks.head(2)
```

```python
#대륙별 국가평균 맥주소비량을:bar chart
```

```python
drinks.groupby('continent')['beer'].mean().plot(kind='bar') #.plot(kind='bar') 
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```

```python
drinks.groupby('continent').mean().plot(kind='bar',stacked=True) #stacled=True
```

```python
# histogram - 구간에 포함된 데이터의 수를 bar
```

```python
drinks['beer'].plot(kind='hist',bins=50)
plt.xlabel('Beer Servings')
plt.ylabel('Frequency')
plt.title('Histogram of Beer')
```

```python
# scatter -> x좌표,y좌표
```

```python
drinks.plot(kind='scatter',x='beer',y='wine') #분포
```
