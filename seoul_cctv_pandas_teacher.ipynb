```python
pop_seoul = pd.read_excel('./data/01. population_in_seoul.xls', usecols = [1,3,6,9,13],
                         skiprows=2)
pop_seoul.head(2)
```

```python
pop_seoul.rename(columns = {'자치구':'구별','계':'인구수','계.1':'한국인',\
                            '계.2':'외국인', '65세이상고령자':'고령자'}, \
                 inplace=True)
```

```python
pop_seoul.head()
```

```python
pop_seoul.drop(26, axis=0, inplace=True)
```

```python
pop_seoul.tail()
```

```python
# 구별 외국인 비율 상위 5개 구...
pop_seoul['외국인비율'] = (pop_seoul['외국인']/pop_seoul['인구수'])*100
pop_seoul.sort_values('외국인비율', ascending=False).head()
```

```python
# 고령자 비율 상위 5개 구...
pop_seoul['고령자비율'] = (pop_seoul['고령자']/pop_seoul['인구수'])*100
pop_seoul.sort_values('고령자비율', ascending=False).head()
```

```python
# 2016년도에 cctv증가율이 가장 높은 구?
# => (2016년도 cctv증가 수) / (2015년도 누적 cctv수)
```

```python
# cctv_seoul['2015년도 누적 cctv수'] = 
cctv_seoul[['2013년도 이전','2014년','2015년']].sum(axis=1)
```

```python
cctv_seoul['2015년도 누적 cctv수'] = cctv_seoul[['2013년도 이전','2014년','2015년']].sum(axis=1)
cctv_seoul.head(2)
```

```python
cctv_seoul['2016년 증가율'] = cctv_seoul['2016년']/cctv_seoul['2015년도 누적 cctv수']
```

```python
cctv_seoul.head()
```

```python
cctv_seoul.sort_values('2016년 증가율', ascending=False)[:5]
```
