광고비: 13,8,10,15,12,14,15,17
매출액: 94,70,90,100,95,100,85,95

1.광고비와 매출액의 평균,분산,표준편차를 각각 구하시오
2.광고비와 매출액의 공분산과 상관계수를 구하시오

```python
import pandas as pd
import numpy as np
```

```python
d = {'광고비':[13,8,10,15,12,14,15,17],
     '매출액':[94,70,90,100,95,100,85,95]}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
#평균

df['광고비_평균'] = df['광고비'].mean()
df['매출액_평균'] = df['매출액'].mean()
```

```python
df
```

```python
#편차

df['광고비_편차'] = df['광고비'] - df['광고비_평균']
df['매출액_편차'] = df['매출액'] - df['매출액_평균']
```

```python
df
```

```python
df['광고비_제곱'] = df['광고비_편차'].apply(lambda x: x**2)
df['매출액_제곱'] = df['매출액_편차'].apply(lambda x: x**2)
```

```python
df['광고비_분산'] = df['광고비_제곱'].mean()
df['매출액_분산'] = df['매출액_제곱'].mean()
```

```python
df
```

```python
#표준편차
```

```python
df['광고비_표준편차'] = df['광고비_분산'].apply(lambda x: np.sqrt(x))
df['매출액_표준편차'] = df['매출액_분산'].apply(lambda x: np.sqrt(x))
```

```python
df
```

```python
#공분산
```

```python
df['광고비_매출액_편차곱'] = df['광고비_편차'] * df['매출액_편차']
```

```python
df['공분산'] = (df['광고비_매출액_편차곱'].sum())/len(df)
```

```python
df
```

```python
#상관관계
```

```python
df['광고비_std'] = df['광고비'].std(ddof=0)
df['매출액_std'] = df['매출액'].std(ddof=0)
```

```python
df['상관관계'] = df['공분산']/(df['광고비_std']*df['매출액_std'])
```

```python
df
```
