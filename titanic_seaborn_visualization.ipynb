```python
#1. pclass의 클래스별 평균연령구하기..
```

```python
df.groupby('pclass')['age'].mean()
```

```python
age_list = []
for i in zip(df.age,df.pclass):
    if np.isnan(i[0]):
        if i[1] == 1:
            age_list.append(38)
        elif [i] == 2:
            age_list.append(29)
        else:
            age_list.append(25)
    else:
        age_list.append(i[0])
age_list
```

```python
df['age'] = age_list
```
